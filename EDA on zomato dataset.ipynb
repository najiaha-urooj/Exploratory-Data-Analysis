# Steps for data analysis

# Step 1: To find out if there are any "Missing values" in the dataset

```python
dataset.isnull().sum().sort_values(ascending=False)
```

```python
#Replacing missing values in cuisine feature by mode 
dataset['Cuisines'] = dataset['Cuisines'].fillna(dataset['Cuisines'].mode()[0])
```

```python
dataset.isnull().sum().sort_values(ascending=False)
```

Now there are no missing values in the dataset

```python
# Importing country code file to extract important feature
dataset_country = pd.read_excel("Country-Code.xlsx")
```

```python
dataset_country.head()
```

```python
dataset.columns
```

We have country code feature in both the files so we are combining the two files to get the country code in the dataset

```python
combine_df = pd.merge(dataset,dataset_country,on = 'Country Code',how = 'left')# merging the files
```

```python
combine_df.head(2)
```

```python
combine_df.columns
```

Now we got a combine dataset including country code feature

```python
#Plotting a pie chart on top 2 countries which uses zomato
country_name = combine_df.Country.value_counts().index
```

```python
country_val = combine_df.Country.value_counts().values
```

```python
plt.pie(country_val[:2],labels = country_name[:2],autopct="%1.1f%%")
plt.show()
```

Observation: Zomato has maximum transaction from India

```python
# Relation of rating  
ratings = combine_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating_count'})
```

```python
ratings
```

```python
import seaborn as sns # importing seaborn library
plt.figure(figsize=(16,9))
sns.barplot(data=ratings, x='Aggregate rating', y='Rating_count', hue='Rating color', palette=sns.color_palette(['pink','red','orange','yellow','green','blue'], as_cmap = True))
plt.show()# White colour is not visible so we took pink
```

```python
plt.figure(figsize=(16,9))
sns.countplot(x='Rating color',data = ratings, palette=sns.color_palette(['pink','red','orange','yellow','green','blue'], as_cmap = True))
plt.show()
```

```python
# find the countries names tht has given 0 ratings
combine_df.groupby(['Aggregate rating','Country']).size().reset_index()
```

Observation: Maximum number of zero ratings is from country India (2139)
